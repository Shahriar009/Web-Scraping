```python
import requests
```

```python
data=[]
for p in range(1,4):
    url = 'https://pharmeasy.in/api/otc/getCategoryProducts?categoryId=717&page={p}&subCategoryId=442'
    r = requests.get(url)
    jason = r.json()
    for i in range(20):
        dic = {
            'Name': jason['data']['products'][i]['name'],
            'Brand': jason['data']['products'][i]['manufacturer'],
            'TypeofProduct': jason['data']['products'][i]['productType'],
            'MrpPrices': jason['data']['products'][i]['mrpDecimal'].split('.')[0] + ' Tk',
            'SalesPrice': jason['data']['products'][i]['salePriceDecimal'].split('.')[0] + ' Tk',
            'DiscountPrice': jason['data']['products'][i]['discountDecimal'].split('.')[0] + ' Tk',
            'Images': jason['data']['products'][i]['images'][0]
        }
        print('----------------------------------------------------')
        print('----------------------------------------------------')
        print('----------------------------------------------------')

        print(dic)

        print('----------------------------------------------------')
        print('----------------------------------------------------')
        print('----------------------------------------------------')
        data.append(dic)
     
    
```

```python
import pandas as pd
df=pd.DataFrame(data)
```
