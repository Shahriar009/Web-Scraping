```python
link_url='https://www.tripadvisor.com/Restaurants-g295424-Dubai_Emirate_of_Dubai.html'
driver.get(link_url)
```

```python
data=[]

for page in range(1,361):
    
    page_url =link_url+ f"-oa{30*(page-1)}"
    driver.get(page_url)
    
    
    for r in range(1,38):
    
        resturent_title=driver.find_element('xpath',f'(//div[@class="tkvCJ u F f Ff K"])[{r}]').text
    
    
        try:
            
            reviews_point=driver.find_element('xpath',f'(//span[@class="IiChw"])[{r}]').text
    
        except:
            
            reviews_point="None"
        
        try:
            
            reviews_link=driver.find_element('xpath',f'(//a[@class="aWhIG _S "])[{r}]').get_attribute('href')
        
        except:
            
            reviews_link="None"
        
        try:
            
            category=driver.find_element('xpath',f'(//div[@class="hBcUX XFrjQ mIBqD"])[{r}]').text
        
        except:
            
            category="None"
        
        try:
            
            comment_link=driver.find_element('xpath',f'(//a[@class="aWhIG _S fnrKq"])[{r}]').get_attribute('href')
        
        except:
            
            comment_link="None"
    
    
        dic={
    
             "resturent_title":resturent_title,
             "reviews_point":reviews_point,
             "reviews_link":reviews_link,
             "category":category,
             "comment_link":comment_link

    
            }
        data.append(dic)
         
    
    driver.find_element('xpath','//a[@class="nav next rndBtn ui_button primary taLnk"]').click()




```
