```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
from selenium.common.exceptions import TimeoutException
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import pandas as pd
```

```python
driver=webdriver.Chrome(ChromeDriverManager().install())
driver
```

```python
link_url='https://www.futurepedia.io/'
driver.get(link_url)
```

```python
for e in range(1,153):
    
    title=driver.find_element('xpath',f'(//h3[@class="MuiTypography-root MuiTypography-h6 MuiTypography-alignLeft css-nxqa8p"])[{e}]').text
    
    try:
        
        review=driver.find_element('xpath',f'(//div[@class="MuiBox-root css-1mpg011"])[{e}]').text
        
    except:
        review="None"
    
    types_for_use=driver.find_element('xpath',f'(//p[@class="MuiTypography-root MuiTypography-body1 MuiTypography-alignLeft ellipsis css-p4n8ke"])[{e}]').text
    
    pricing=driver.find_element('xpath',f'(//div[@class="MuiChip-root MuiChip-outlined MuiChip-sizeMedium MuiChip-colorDefault MuiChip-outlinedDefault css-1e478s8"])[{e}]').text
    
    try:
        per_month=driver.find_element('xpath',f'(//p[@class="MuiTypography-root MuiTypography-body1 css-1qm6pqb"])[{e}]').text
        
    except:
        per_month="None"
        
    
    categories=driver.find_element('xpath',f'(//div[@class="css-yt9ay8"])[{e}]').text
    
    site_url=driver.find_element('xpath',f'(//a[@class="MuiButtonBase-root MuiButton-root MuiButton-contained MuiButton-containedPrimary MuiButton-sizeMedium MuiButton-containedSizeMedium MuiButton-disableElevation MuiButton-root MuiButton-contained MuiButton-containedPrimary MuiButton-sizeMedium MuiButton-containedSizeMedium MuiButton-disableElevation css-p9t5dc"])[{e}]').get_attribute('href')
    
    try:
        features=driver.find_element('xpath',f'(//div[@style="display: flex; justify-content: space-between; align-items: center; margin-right: 25px; user-select: none;"])[{e}]').text
        
    except:
        features="None"
    d={
    "title":title,
        
    "review":review,
    "types_for_use":types_for_use,
    "pricing":pricing,
    "per_month":per_month,
    "categories":categories,
    "site_url":site_url,
    "features":features
    
    
    }
    
    print(d)
```
